# Are there any states that are particularly worse or better off when considering their population? (How prevalent is the disease in the India)
#### -To determine if states are particularly worse or better off considering their population,you can analyze the number of cases relative to their population. One way to achieve this is by calculating the number of cases per capita (e.g., cases per 1,000 or 100,000 people) for each state and then determining which states have the highest and lowest values.

```python
df["Cases_Per_100K"] = round(((df["Total_Cases"]/ df["Population"])* 100000))
```

# How does the total number of cases, active cases, discharges, and deaths correlate with the population of each state?

```python
sns.heatmap(df_corr, annot=True)
```

- Total cases vs population has a pearson correlation coefficient of -0.067.

- Active cases vs population has a pearson correlation coefficient of -0.054

- Discharged cases vs population has a pearson correlation coefficient of -0.066

- Number of deaths vs population has a pearson correlation coefficient of -0.12

Let's interpret these Pearson correlation coefficients:

1. **Total cases vs. population: -0.067**
    - This indicates a very weak negative linear relationship between the total number of cases and the population. The negative sign implies that, on average, as one variable increases, the other tends to decrease slightly. However, given how close the value is to 0, the relationship is weak, and it suggests that the total number of cases is largely independent of the state's population size.

2. **Active cases vs. population: -0.054**
    - This correlation coefficient also indicates a very weak negative linear relationship between the number of active cases and the population. Like the previous metric, the relationship is close to non-existent, and there's very little dependency of active cases on the state's population size.

3. **Discharged cases vs. population: -0.066**
    - This coefficient is similar to that of total cases vs. population. There's a very weak negative correlation between the number of discharged cases and the population. This suggests that the number of people discharged after recovering from COVID-19 in a state doesn't have a strong dependency on the state's population.

4. **Number of deaths vs. population: -0.12**
    - This indicates a weak negative linear relationship between the number of deaths and the population size. Though this coefficient is slightly more pronounced than the others, it's still relatively close to 0. This implies that the number of deaths in a state due to COVID-19 is somewhat, though not strongly, independent of the population size.

**General Interpretation:**
All the correlation coefficients are negative, suggesting that states with larger populations do not necessarily have proportionally larger counts of cases, active cases, discharges, or deaths. However, it's essential to note that all the correlation values are very close to zero, indicating very weak relationships. This means that other factors, beyond just population size, play a significant role in these COVID-19 metrics for each state.

Additionally, correlation does not imply causation. While these metrics suggest a lack of strong linear relationship between population size and the COVID-19 metrics, it doesn't provide information on the causal factors or other potential relationships that might exist in a non-linear manner.

# Which states have the highest number of cases, discharges, and deaths per capita?

```python
df["Discharged_Per_100K"] = round(((df["Discharged"]/ df["Population"])* 100000))

df["Deaths_Per_100K"] = round(((df["Deaths"]/ df["Population"])* 100000))
```

```python
df["Discharged_Per_100K"]
```

```python
#Dicharged per 100K column
Discharged_Per_100K_Column = df["Discharged_Per_100K"]

#Select the row/rows in the dataframe or data from the Discharged_Per_100K_Column which has the highest Discharged per capita (100K)
highest_discharges_per_capita = df[Discharged_Per_100K_Column == Discharged_Per_100K_Column.max()]

#Give only the state column and the exact state/region with the highest discharges per 100K
highest_discharges_per_capita["State"].iloc[0] #iloc[0] selects only the first row or position index- leaves out the index
```

- Maharashtra is the state with the highest number of discharged patients per capita (100K)

```python
#Deaths per 100K column
Deaths_Per_100K_Column = df["Deaths_Per_100K"]

#Select the row/rows from the Deaths_Per_100K_Column with the highest deaths per capita
highest_deaths_per_capita = df[Deaths_Per_100K_Column == df["Deaths_Per_100K"].max()]

#Give only the state (not the index) and the exact state/region with the highest deaths per 100k
highest_deaths_per_capita["State"].iloc[0]
```
